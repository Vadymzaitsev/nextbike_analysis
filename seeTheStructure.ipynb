Let's check out the structure of the server responce to our query.
https://api.nextbike.net/maps/nextbike-live.json?city=685

```python
import pandas as pd
from datetime import datetime

url = "https://api.nextbike.net/maps/nextbike-live.json?city=685"
responce = pd.read_json(url)

dictionary = responce['countries'].to_dict()
dictionary
```

```python
useless_keys=['lat', 'lng', 'zoom', 'name', 'hotline', 'domain', 
              'language', 'email','timezone', 'currency', 'country_calling_code',
              'system_operator_address', 'country', 'country_name', 'terms', 
              'policy', 'website', 'show_bike_types', 'show_bike_type_groups', 'show_free_racks'
              'capped_available_bikes', 'no_registration', 'pricing', 'vat', 'faq_url', 'store_uri_android', 
              'store_uri_ios']
for i in useless_keys:
    if i in dictionary[0]:
        removed_value = dictionary[0].pop(i)
        print(f"Removed key '{i}' with value: {removed_value}")

dictionary[0]

```
